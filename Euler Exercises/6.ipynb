```python
# Initialize variables
sum_of_squares = 0
sum_of_numbers = 0

# Loop through the numbers from 1 to 100
for i in range(1, 101):
    sum_of_squares += i ** 2  # Add the square of i to sum_of_squares
    sum_of_numbers += i  # Add i to sum_of_numbers

# Compute the square of the sum
square_of_sum = sum_of_numbers ** 2

# Calculate the difference
difference = square_of_sum - sum_of_squares

# Print the result
print(difference)
```

#### Efficiency:
